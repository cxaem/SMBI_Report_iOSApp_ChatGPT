# Introduction

## Problem to Solve

The central question that this project aims to tackle is: **"How did the introduction of chat history management features in the ChatGPT App (release 1.2023.23) influence user sentiment as seen in the App Store reviews available until June 25, 2023?"** 

## Hypothesis

In light of the rising data privacy and management concerns, we anticipate that the introduction of **chat history management** features in release **1.2023.23** would have been received positively, resulting in a noticeable shift in user sentiment in reviews related to data privacy.

During the data processing phase, we adopted a methodical approach for data cleansing and preparation. We designed a preprocess() function that carries out a suite of preprocessing techniques. These techniques include the removal of punctuation and HTML markup, along with the application of stemming and lemmatization procedures. Our objective was to refine the dataset and prime it for more sophisticated analyses, modeling, and predictive tasks. This systematic preprocessing phase was a pivotal step in ensuring the accuracy and reliability of our subsequent data investigations.

```python
# Define the preprocessing function
def preprocess(sentence):
    sentence = str(sentence).lower()
    sentence = sentence.replace('{html}', "")
    sentence = re.sub('<.*?>', '', sentence)  # remove HTML tags
    sentence = re.sub(r'http\S+', '', sentence)  # remove URLs
    sentence = re.sub('[0-9]+', '', sentence)  # remove numbers
    tokens = RegexpTokenizer(r'\w+').tokenize(sentence)  # tokenize the sentence
    filtered_words = [w for w in tokens if len(w) > 2 and w not in stopwords.words('english')]
    lemma_words = [WordNetLemmatizer().lemmatize(w) for w in filtered_words]
    return " ".join(lemma_words)

# Concatenate 'title' and 'content' for text analysis 
reviews_df['review_clean'] = reviews_df['title'] + " " + reviews_df['review']

# Apply the function to the 'review' column
reviews_df['review_clean'] = reviews_df['review_clean'].apply(preprocess)

# Create feature 'numwords' to capture the number of words each 'review_clean' has 
reviews_df['numwords'] = reviews_df['review_clean'].apply(lambda x: len(x.split()))

# Display the first few rows of the dataframe
reviews_df.head()
```

We introduce a new column labeled 'numwords' to our dataset. This column measures the quantity of words within each entry found in the 'review_clean' column. This newly added feature can provide significant insights for future analyses by serving as a metric for review length

#### Textblob

In order to assess the impact of the release on both the frequency of keywords and the sentiment of the reviews, it is necessary to determine the sentiment expressed in all the reviews. To accomplish this, we have various tools at our disposal. 

Given that the dataset consists of reviews, TextBlob emerges as a reasonable choice for sentiment analysis. TextBlob provides a simplified approach to sentiment analysis with polarity scores (negative, neutral, positive) and subjectivity scores, making it suitable for gaining a general understanding of sentiment. 

```python
# Function to calculate polarity and subjectivity
def calc_polarity_subjectivity(df, text_column):
    df['polarity_txblb'] = df[text_column].apply(lambda x: TextBlob(x).sentiment.polarity)
    df['subjectivity_txblb'] = df[text_column].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Function to get sentiment
def get_sentiment(polarity):
    if polarity >= 0.05:
        return 'Positive'
    elif polarity <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Function to assign sentiment categories
def assign_sentiment_categories(df, sentiment_column):
    df['is_review_positive'] = df[sentiment_column].apply(lambda x: 1 if x == 'Positive' else 0)
    df['is_review_neutral'] = df[sentiment_column].apply(lambda x: 1 if x == 'Neutral' else 0)
    df['is_review_negative'] = df[sentiment_column].apply(lambda x: 1 if x == 'Negative' else 0)

# Calculate polarity and subjectivity for each review
calc_polarity_subjectivity(reviews_df, 'review_clean')

# Assign sentiment based on polarity
reviews_df['sentiment_txblb'] = reviews_df['polarity_txblb'].apply(get_sentiment)

# Assign sentiment categories
assign_sentiment_categories(reviews_df, 'sentiment_txblb')
```

```python
# Create a crosstab of sentiments and ratings
sentiments_ratings_ct = pd.crosstab(reviews_df['rating'],reviews_df['sentiment_txblb'])

# Display the crosstab
sentiments_ratings_ct
```

```python
def generate_crosstab(df, release_column, rating_column, sentiment_column):
    # Separate the dataset into two: before and after the release
    df_before = df[df[release_column] == 0]
    df_after = df[df[release_column] == 1]

    # Calculate the cross tabulations
    crosstab_before = pd.crosstab(df_before[rating_column], df_before[sentiment_column])
    crosstab_after = pd.crosstab(df_after[rating_column], df_after[sentiment_column])

    return crosstab_before, crosstab_after

def plot_heatmaps(crosstab_before, crosstab_after, ax):
    # Create a heatmap for reviews before the release
    sns.heatmap(crosstab_before, annot=True, fmt="d", cmap="YlGnBu", ax=ax[0])
    ax[0].set_title('Sentiments for Reviews Before the Release')

    # Create a heatmap for reviews after the release
    sns.heatmap(crosstab_after, annot=True, fmt="d", cmap="YlGnBu", ax=ax[1])
    ax[1].set_title('Sentiments for Reviews After the Release')

# Set up the matplotlib figure
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Generate crosstabs
crosstab_before, crosstab_after = generate_crosstab(reviews_df, 'is_review_after_release_23', 'rating', 'sentiment_txblb')

# Plot heatmaps
plot_heatmaps(crosstab_before, crosstab_after, axs)

# Display the plots
plt.tight_layout()
plt.show()
```

The two heatmaps above show the cross-tabulation of sentiments and ratings for reviews from before and after the release. Looking at the structure of the tables, we can see a similar pattern in both periods. 

This indicates that higher ratings are often associated with positive sentiments, as one would logically expect. The tables also show that negative sentiments are more frequent in reviews with a rating of 1 in both periods. This indicates that reviewers who gave the lowest rating expressed negative sentiments in their reviews.

While it is important to note that the absolute numbers vary while the distribution patterns are similar, the sentiment-ratings relationship appears consistent before and after the release. This suggests that the datasets from both periods are comparable regarding sentiment distribution across different ratings.

## Keyword Analysis

The keyword analysis provides insights into how often certain keywords related to privacy and history management were mentioned in the reviews before and after the release of the new feature.

```python
def compute_word_freqs(df, release_column, review_column, word_list):
    # Split the reviews before and after the release into individual words
    before_release_reviews = df[df[release_column] == 0][review_column]
    before_release_text = " ".join(before_release_reviews)
    before_release_words = before_release_text.split()

    after_release_reviews = df[df[release_column] == 1][review_column]
    after_release_text = " ".join(after_release_reviews)
    after_release_words = after_release_text.split()

    # Create a DataFrame to store the word frequencies
    word_freq_df = pd.DataFrame(columns=['Keyword', 'Before Release', 'After Release', 'Percentage Change (%)'])

    # Calculate the word frequencies for the "before the release" data
    before_release_word_freq = {word: before_release_words.count(word) for word in word_list}

    # Calculate the word frequencies for the "after the release" data
    after_release_word_freq = {word: after_release_words.count(word) for word in word_list}

    # Fill the DataFrame with the word frequencies
    for word in word_list:
        word_freq_df = word_freq_df.append({
            'Keyword': word,
            'Before Release': before_release_word_freq[word],
            'After Release': after_release_word_freq[word],
            'Percentage Change (%)': round((((after_release_word_freq[word] - before_release_word_freq[word]) / before_release_word_freq[word] * 100)),2)
        }, ignore_index=True)

    return word_freq_df

# Define the keywords related to privacy and history management
keywords = ['privacy', 'data', 'history', 'management', 'security', 'control']

# Compute word frequencies
keyword_freq_df = compute_word_freqs(reviews_df, 'is_review_after_release_23', 'review_clean', keywords)

# Display the keyword frequency table
keyword_freq_df
```

### Findings:

* An 80% decrease in the term "privacy" and a 62% decrease in "data" suggest lessened user discussions around these topics after the new feature was introduced.

* "History" related to the new feature decreased by 31%, indicating less frequent discussions post-release.

* "Management" was not mentioned at all after the release, suggesting user concerns about management decreased.

* "Security" and "control" also saw significant decreases of about 81% and 64% respectively.

These results imply that the new chat history management feature may have addressed user concerns about privacy, data, history management, security, and control. Next, we'll conduct a sentiment analysis to understand the emotional reaction of users towards this new feature.

```python
def plot_sentiment_distributions(df, ax, title):
    # Plot the sentiment distribution
    sns.countplot(x='is_review_after_release_23', hue='sentiment_txblb', data=df, ax=ax)
    ax.set_title(title)
    ax.set_xticklabels(['Before', 'After'])

# Create a figure for subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Plot sentiment distributions
plot_sentiment_distributions(reviews_df, axs[0], 'All Reviews: Sentiment Distribution Before and After Release')
plot_sentiment_distributions(reviews_with_keywords_df, axs[1], 'Keyword Reviews: Sentiment Distribution Before and After Release')

# Adjust the layout and display the plots
plt.tight_layout()
plt.show()
```

```python
def calc_sentiment_change(df, name):
    # Calculate sentiment distribution before and after release
    sentiment_distribution = df.groupby(['is_review_after_release_23', 'sentiment_txblb']).size()

    # Calculate proportions of sentiments
    sentiment_proportion = sentiment_distribution.groupby(level=0).apply(lambda x: x / x.sum())

    # Print change in sentiment distribution and proportions
    for sentiment in ['Negative', 'Neutral', 'Positive']:
        print(f"Change in {sentiment.lower()} sentiment for {name}: {sentiment_distribution.loc[1, sentiment] - sentiment_distribution.loc[0, sentiment]}")
        print(f"Change in proportion of {sentiment.lower()} sentiment for {name}: {sentiment_proportion.loc[1, sentiment] - sentiment_proportion.loc[0, sentiment]}")
    print()

# Call the function for each df
for df, name in [(reviews_df, 'all reviews'), (reviews_with_keywords_df, 'reviews with keywords')]:
    calc_sentiment_change(df, name)
```

* Negative and neutral sentiments increased by around 5% and 2%, respectively, while positive sentiment fell by approximately 8% post-release.

* Reviews with keywords related to privacy and history management showed an 11% increase in negative sentiments, an 8% increase in neutral sentiments, and a 19% decrease in positive sentiments.

```python
def calc_change_and_ttest(df, column):
    # Calculate average values before and after release
    avg_values = df.groupby('is_review_after_release_23')[column].mean()

    # Print change in average values
    print(f"Change in average {column} for all reviews: {avg_values.loc[1] - avg_values.loc[0]}")

    # Separate the scores before and after the feature release
    before_release = df[df['is_review_after_release_23'] == 0][column]
    after_release = df[df['is_review_after_release_23'] == 1][column]

    # Perform the t-test
    t_stat, p_value = stats.ttest_ind(before_release, after_release)
    
    print(f"{column.capitalize()} T-test:")
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")
    print()

# Call the function for each df and column
for df, name in [(reviews_df, 'all reviews'), (reviews_with_keywords_df, 'reviews with keywords')]:
    print(f"For {name}:")
    for column in ['polarity_txblb', 'subjectivity_txblb']:
        calc_change_and_ttest(df, column)
```

### Findings

* The average polarity declined by approximately 0.035 post-release, suggesting a slight shift towards more negative sentiments.
* The average subjectivity decreased by around 0.030, indicating a shift towards more objective language in user reviews.
* Reviews with keywords related to privacy and history management showed a larger decrease in polarity and subjectivity.

The findings suggest that the new feature might have prompted more objective discussions but was received with more negativity than expected, especially in reviews addressing privacy and history management aspects. ## Rating Analysis

```python
def visualize_ratings(df, name, ax):
    # Plot the distribution of ratings before and after the release
    sns.countplot(x='rating', hue='is_review_after_release_23', data=df, ax=ax)
    ax.set_title(name)
    ax.set_xlabel('Rating')
    ax.set_ylabel('Count')
    ax.legend(title='Review Time', title_fontsize='13', labels=['Before Release', 'After Release'])

# Prepare the subplots
fig, axs = plt.subplots(1, 2, figsize=(20, 5))

# Call the function for each df and axis
visualize_ratings(reviews_df, 'All Reviews', axs[0])
visualize_ratings(reviews_with_keywords_df, 'Reviews with Keywords', axs[1])

# Adjust the layout and display the plots
plt.tight_layout()
plt.show()
```

```python
def analyze_ratings(df, name):
    # Calculate average ratings
    avg_rating = df.groupby('is_review_after_release_23')['rating'].mean()

    # Print average ratings
    print(f"Average rating for {name} (Before Release): {avg_rating[0]:.2f}")
    print(f"Average rating for {name} (After Release): {avg_rating[1]:.2f}")

    # Calculate the count of each rating before and after release
    rating_counts = df.groupby(['is_review_after_release_23', 'rating']).size().unstack(fill_value=0)

    # Calculate and print the nominal drop in count for each rating
    for rating in rating_counts.columns:
        drop_reviews = rating_counts.loc[0, rating] - rating_counts.loc[1, rating]
        print(f"Drop in count for rating {rating} for {name}: {drop_reviews}")

    # Perform the t-test
    before_release = df[df['is_review_after_release_23'] == 0]['rating']
    after_release = df[df['is_review_after_release_23'] == 1]['rating']
    t_stat, p_value = stats.ttest_ind(before_release, after_release)

    # Print t-test results
    print(f"T-test results for {name}:")
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")
    print()

# Call the function for each df and axis
analyze_ratings(reviews_df, 'All Reviews')
analyze_ratings(reviews_with_keywords_df, 'Reviews with Keywords')
```

### Findings

* The average rating declined from 3.88 to 3.47 post-release, indicating a shift towards lower ratings.

* The count of each rating from 2 to 5 decreased, with the most significant drop observed for the highest rating of 5.

* Reviews with keywords related to privacy and history management showed a more substantial drop in the average rating and a decrease in the count of higher ratings.
